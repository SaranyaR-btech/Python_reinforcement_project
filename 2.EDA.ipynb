```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('house_clean.csv')
```

                                                        UNIVARIATE ANALYSIS

```python
sns.distplot(df['price'])
```

```python
sns.distplot(df['sqft_living'])
```

```python
df['bedrooms'].value_counts().plot(kind='bar')
```

```python
df['condition'].value_counts().plot(kind='bar')
```

```python
df['bathrooms'].value_counts().plot(kind='bar')
```

                                                             BIVARIATE ANALYSIS

```python
df.plot.hexbin(x='price',y='sqft_living',gridsize=15)
```

```python
df.plot.hexbin(x='bedrooms',y='price',gridsize=15)
```

```python
avg_price=df.groupby('view')['price'].mean()

avg_price.plot(kind='bar')
plt.xticks([0, 1,2,3,4],rotation=0)
plt.show()
```

```python
avg_price=df.groupby('waterfront')['price'].mean()


avg_price.plot(kind='bar')
plt.ylabel("price")
plt.xticks([0, 1],rotation=0)
plt.show()
```

                                             MULTIVARIATE ANALYSIS

```python
df['day']=df['day'].astype(str)
df['year']=df['year'].astype(str)
df['month']=df['month'].astype(str)
num_cols = df.select_dtypes(include=['number'])
corr_matrix = num_cols.corr()

plt.figure(figsize=(18,12))

sns.heatmap(corr_matrix, annot=True, fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```
